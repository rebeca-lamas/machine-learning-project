```python
import pandas as pd
```

```python
batch_1 = pd.read_csv('../data/2017_data_batch_1.csv')
batch_2 = pd.read_csv('../data/2017_data_batch_2.csv')
batch_3 = pd.read_csv('../data/2017_data_batch_3.csv')
batch_4 = pd.read_csv('../data/2017_data_batch_4.csv')
```

```python
df = pd.DataFrame()
```

```python
df = pd.concat([batch_1, batch_2, batch_3, batch_4], ignore_index=True)
```

```python
len(df)
```

```python
df.head(100)
```

```python
df = df[df.box_office.notnull()]
```

```python
len(df)
```

```python
df.to_csv('../data/cleaned_2017_data.csv', index=False)
```

```python
test = df[['title', 'box_office', 'country', 'genre', 'rating', 'runtime', 'production']]
```

```python
test
```

```python
test['country'] = test['country'].str.count(',')
test['country'] = test['country']+1
test.head()
```

```python
test['genre']= test['genre'].str.count(',')
test['genre']= test['genre']+1
test.head()
```

```python
test['rating'].value_counts()
```

```python
test['rating']= test['rating'].str.replace('Not Rated', 'Unrated')
test.head()
```

```python
test['rating']=test['rating'].fillna('Unrated')
test.head()
```

```python
test['runtime'] = test['runtime'].astype(str)
```

```python
test['runtime'] = test['runtime'].apply(lambda x: x.replace(' min', ''))
test.head()
```

```python
test['runtime'] = test['runtime'].apply(lambda x: x.replace('nan', '0'))
test.head()
```

```python
test['runtime']= test['runtime'].astype(int)
```

```python
test['production']= test['production'].fillna('NaN')
test.head()
```

```python
test['box_office'] = test['box_office'].apply(lambda x: x.replace('$', ''))
test.head()
```

```python
test['box_office'] = test['box_office'].apply(lambda x: x.replace(',', ''))
test.head()
```

```python
test['box_office']= test['box_office'].astype(int)
```

```python
test.dtypes
```

```python
test['year'] = 2017
```

```python
test.to_csv('../data/test_data_2017.csv', index=False)
```
