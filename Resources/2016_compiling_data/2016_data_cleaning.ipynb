```python
import pandas as pd
```

```python
batch_1 = pd.read_csv('../data/2016_data_batch_1.csv')
batch_2 = pd.read_csv('../data/2016_data_batch_2.csv')
batch_3 = pd.read_csv('../data/2016_data_batch_3.csv')
batch_4 = pd.read_csv('../data/2016_data_batch_4.csv')
```

```python
df = pd.DataFrame()
```

```python
df = pd.concat([batch_1, batch_2, batch_3, batch_4], ignore_index=True)
```

```python
len(df)
```

```python
df.head(100)
```

```python
total = len(df.country.unique())
unique = df.country.unique()
for i in range (total):
    print(unique[i])
```

```python
df.country = df.country.str.lower()
```

```python
df = df[df['country'].str.contains("usa", na=False)]
```

```python
total = len(df.country.unique())
unique = df.country.unique()
for i in range (total):
    print(unique[i])
```

```python
len(df)
```

```python
df = df[df.box_office.notnull()]
```

```python
len(df)
```

```python
# drop min number of non-NA values
df = df.dropna(thresh=7)
df.head(100)
```

```python
len(df)
```

```python
df.to_csv('../data/compiled_2016_data.csv', index=False)
```

```python
train = df[['title', 'box_office', 'country', 'genre', 'production', 'rating', 'runtime']]
train.head()
```

```python
train = train.reset_index(drop = True)
train.head()
```

```python
train['country'] = train['country'].str.count(',')
train['country'] = train['country'] + 1
train.head()
```

```python
train['genre'] = train['genre'].str.count(',')
train['genre'] = train['genre'] + 1
train.head()
```

```python
train['rating'].value_counts()
```

```python
train['rating'] = train['rating'].str.replace('Not Rated', 'Unrated')
train.head()
```

```python
train['rating'] = train['rating'].fillna('Unrated')
```

```python
train['rating'].value_counts()
```

```python
train['runtime'] = train['runtime'].astype(str)
```

```python
train['runtime'] = train['runtime'].apply(lambda x: x.replace(' min',''))
train.head()
```

```python
train['runtime'] = train['runtime'].apply(lambda x: x.replace('nan','0'))
train.head()
```

```python
train['runtime'] = train['runtime'].astype(int)
```

```python
train.production.value_counts()
```

```python
train['production'] = train['production'].fillna('NaN')
```

```python
train['box_office'] = train['box_office'].apply(lambda x: x.replace('$',''))
train.head()
```

```python
train['box_office'] = train['box_office'].apply(lambda x: x.replace(',',''))
train.head()
```

```python
train['box_office'] = train['box_office'].astype(int)
train.head()
```
